```python
# tips dataset

df = sns.load_dataset('tips')
df.head()
```

```python
sns.pairplot(df, corner=True, hue='day')
```

```python
df.describe().round(2)
```

```python
df[['total_bill','tip','size']].cov().round(2)


# --------------------

columns = ['total_bill','tip','size']

covariance = df[columns].cov()

covariance.round(2)
```

##### Covariance Matrix

- total bill vs total bill = 79.25
- 

##### Correlation

```python
covariance.round(2)
sns.heatmap(covariance, cmap= 'rocket', annot=True, fmt=".2f", vmin= 0, vmax=1 )
```

```python
corelation = df[columns].corr(method= 'pearson') # corelation matrix - spearman - kendall - 
corelation.round(2)
```
